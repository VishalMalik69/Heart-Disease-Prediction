```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import sklearn 
```

```python
df = pd.read_csv("heart_disease_uci.csv")
df.head()
```

```python
df.info()
```

```python
df = df.drop(columns = ["id" , "ca"])
```

```python
df.head()
```

```python
df = pd.get_dummies(df , drop_first = True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.fillna(df.median(numeric_only = True) , inplace = True )
df.info()
```

```python
df = df.replace({True: 1, False: 0})
df["num"] = df["num"].replace({1: 1, 2: 1, 3: 1, 4: 1})
```

```python
df
```

```python
df.info()
```

```python
X = df.drop(columns = ["num"])
y = df["num"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import RobustScaler
from sklearn.tree import DecisionTreeClassifier 
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score , precision_score  , recall_score , f1_score , classification_report , confusion_matrix
```

```python
X_test , X_train , y_test , y_train = train_test_split(X,y , random_state = 42 , test_size = 0.2)
```

```python
scaler = RobustScaler()
```

```python
X_train[["age" , "trestbps" , "chol" , "thalch" , "oldpeak"]] = scaler.fit_transform(X_train[["age" , "trestbps" , "chol" , "thalch" , "oldpeak"]])
X_test [["age" , "trestbps" , "chol" , "thalch" , "oldpeak"]] = scaler.transform(X_test [["age" , "trestbps" , "chol" , "thalch" , "oldpeak"]])
```

```python
LogisticRegression = LogisticRegression (max_iter = 1000 , random_state = 42 , class_weight = "balanced" )
LogisticRegression.fit(X_train , y_train)
y_predict = LogisticRegression.predict(X_test)
```

```python
accuracy = accuracy_score(y_test , y_predict)
precision = precision_score(y_test , y_predict)
recall = recall_score(y_test , y_predict)
f1 = f1_score(y_test , y_predict)
classification = classification_report(y_test , y_predict)
matrix = confusion_matrix(y_test , y_predict)
```

```python
print(df["num"].value_counts())
```

```python
print("=" * 50)
print("Model Evaluation Metrics")
print("=" * 50)

print(f"Accuracy  : {accuracy:.4f}")
print(f"Precision : {precision:.4f}")
print(f"Recall    : {recall:.4f}")
print(f"F1 Score  : {f1:.4f}")

print("\n" + "=" * 50)
print("Classification Report")
print("=" * 50)
print(classification)

print("=" * 50)
print("Confusion Matrix")
print("=" * 50)
print(matrix)
```
